# Step 4 | Univariate Analysis

## Step 4.1 | Categorical Variables Univariate Analysis

```python
# Create the subplots
fig = make_subplots(rows=3, cols=3, specs=[[{'type':'domain'}]*3]*3, vertical_spacing=0.05, horizontal_spacing=0.01)

# Loop through all the features and add the pie chart to the subplot
for i, feature in enumerate(df_categorical.columns):
    value_counts = df_categorical[feature].value_counts()
    labels = value_counts.index.tolist()
    values = value_counts.values.tolist()

    # Define color map based on orangered color
    cmap = colors.LinearSegmentedColormap.from_list("slateblue", ["slateblue", "white"])
    norm = colors.Normalize(vmin=0, vmax=len(labels))
    color_list = [colors.rgb2hex(cmap(norm(i))) for i in range(len(labels))]

    # Create the pie chart
    pie_chart = go.Pie(
        labels=labels,
        values=values,
        hole=0.6,
        marker=dict(colors=color_list, line=dict(color='white', width=3)),
        textposition='inside',
        textinfo='percent+label',
        title=feature,  # Add title with the feature name
        title_font=dict(size=25, color='black', family='Calibri')
    )

    # Add the pie chart to the subplot
    if i < 8:
        row = i // 3 + 1
        col = i % 3 + 1
        fig.add_trace(pie_chart, row=row, col=col)

# Update the layout
fig.update_layout(showlegend=False, height=1000, width=980, 
                   title={
                          'text':"Distribution of Categorical Variables",
                          'y':0.95,
                          'x':0.5,
                          'xanchor':'center',
                          'yanchor':'top',
                          'font': {'size':28, 'color':'black', 'family':'Calibri'}
                         })

# Show the plot
fig.show()
```

- The dataset is evenly divided across 4 price_range classes, making it balanced.
- Mobile phones have nearly identical frequencies regarding features like Bluetooth, 4G, touch screen, Wifi, dual SIM support, and processing cores.
- Approximately 76% of the phones support 3G.

**Strategy**:

- Plan to implement Support Vector Machine (SVM), which is less sensitive to noise and outliers compared to algorithms like decision trees.
- Based on the historical context, set the minimum acceptable value of 5 for px_height and 2 for sc_w.
- Identify the index of noise samples but delay action to clean or impute them until working with noise-sensitive models like decision trees.

```python
# Detect Noise values in sc_w 
sc_w_NoiseIndex = df[df['sc_w'] < 2].index

# Detect Noise values in px_height
px_height_NoiseIndex = df[df['px_height'] < 5].index
```

# Step 5.4 | Feature Selection

### Definition:

Feature selection is choosing a relevant subset of features from a larger set in a dataset to enhance a machine learning model's performance. It helps in accurate prediction of the target variable by excluding irrelevant or redundant features.

### Advantages:

- **Improves Accuracy**: Reduces overfitting, where a model incorrectly captures noise.
- **Efficiency**: Reduces time and computational resources by eliminating unnecessary features.
- **Enhances Interpretability**: Identifies crucial features, making the model more understandable.

### Techniques:

- **Filter Methods**: Rank features using statistical measures to gauge correlation with the target variable (e.g. Pearson correlation coefficient, chi-squared test)
- **Wrapper Methods**: Evaluate performance on various feature subsets, selecting the best-performing subset(e.g. Drop-column feature selection)
- **Embedded Methods**: Include feature selection during model training, focusing on the features contributing to performance (e.g. LASSO regression, Ridge regression, Decision trees)
- **Dimensionality Reduction**: Transform original features into a lower-dimensional space, preserving variance (e.g.PCA Principal Component Analysis)

### Strategy:
- Among the filter methods, we will perform the Pearson correlation coefficient.
- We will perform Drop-column Feature Importance as a Wrapper method.
- We will implement models such as Decision Tree and Random Forest as Embedded methods.

```python
# Define a colormap
orangered_cmap = LinearSegmentedColormap.from_list('slateblue', ['#FFECEC','slateblue'])

# Calculation of the Pearson correlation
target = 'price_range'
df_ordered = pd.concat([df.drop(target,axis=1), df[target]],axis=1)
corr = df_ordered.corr()

# Create a mask so that we see the correlation values only once
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask,1)] = True

# Plot the heatmap correlation
plt.figure(figsize=(18,12))
sns.heatmap(corr, mask=mask, annot=True, cmap=orangered_cmap, fmt='.2f', linewidths=0.2)
plt.show()
```

- price_range is highly correlated with ram (Ï = 0.92).
- price_range has a low correlation value with the rest of the features, but this cannot be used as a criterion to remove these features since the pearson correlation only expresses the linear relationship between two variables.
